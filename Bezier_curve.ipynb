```python
# Importando bibliotecas

import numpy as np
import matplotlib.pyplot as plt
import math
```



---






# Definição

Para qualquer n inteiro não-negativo e $P_0,  P_1, ..., P_n$ pontos distintos, a curva de Bézier de ordem n é dada pela equação:

$ B_n(t, P_0, ...,P_n) = \sum_{k=0}^{n}\binom{n}{k}(1-t)^{n-k}t^k \cdot P_k $,

onde $0\leq t \leq 1$.

Por exemplo, quando n = 5:

$B_n(t, P_0, ...,P_5) =  (1 - t)^5P_0 + 5t(1 - t)^4P_1 + 10t^2(1 - t)^3 P_2 + 10t^3 (1-t)^2 P_3 + 5t^4(1-t)P_4 + t^5 P_5 $, 

$0\leq t \leq 1$.

# Curva de Bézier em 3 dimensões

```python
# Funcao para calcular valores e plotar o a curva 

def plot_bezier_2D(pontos):
    x = [ponto[0] for ponto in pontos] # vetor com coordenadas x[t]
    y = [ponto[1] for ponto in pontos] # vetor com coordenadas y[t]

    fig, ax = plt.subplots()  

    N = np.size(x)-1 # = Grau da curva = qtd de pontos - 1, pois índice deve começar em 0
    t = np.linspace(0, 1, 500) 

    def binNewton(i,n,t) :
        return ((math.factorial(n)/(math.factorial(i)*math.factorial(n-i)))*pow(1-t,n-i)*pow(t,i))

    def somatorio(vetorCoordenada, t):
        soma = 0
        for i in range(0,N+1):
            soma = soma + binNewton(i,N,t)*vetorCoordenada[i]
        return soma

    B_x = somatorio(x, t)
    B_y = somatorio(y, t)


    ax.plot(B_x, B_y, '-') 
    plt.plot(x,y,'o')
    ax.grid(True)
    ax.set_xlabel("eixo x", fontsize=14)
    ax.set_ylabel("eixo y", fontsize=14)

    plt.plot()
```

## Exemplo de pontos de controle ($P_0,  P_1, ..., P_n$)

```python
pontos = [[5, 0], [0, 2], [0, 10], [6, 10], [14, -2], [-4, -2], [4, 10], [10, 10], [10, 2], [5, 0]]
plot_bezier_2D(pontos)
```

# Curva de Bézier em 3 dimensões

```python
# Funcao para calcular valores e plotar o a curva 

def plot_bezier_3D(pontos):
    # Criando a figura e projeção em 3D
    fig = plt.figure()
    ax = fig.gca(projection='3d')

    x = [ponto[0] for ponto in pontos] 
    y = [ponto[1] for ponto in pontos] 
    z = [ponto[2] for ponto in pontos] 
    

    N = np.size(x)-1 # = Grau da curva = qtd de pontos - 1, pois índice deve começar por 0
    t = np.linspace(0, 1, 500) 


    def binNewton(i,n,t) :
        return ((math.factorial(n)/(math.factorial(i)*math.factorial(n-i)))*pow(1-t,n-i)*pow(t,i))

    def somatorio(vetorCoordenada, t):
        soma = 0
        for i in range(0,N+1):
            soma = soma + binNewton(i,N,t)*vetorCoordenada[i]
        return soma

    B_x = somatorio(x, t)
    B_y = somatorio(y, t)
    B_z = somatorio(z, t)


    ax.plot(B_x, B_y, B_z)
    plt.plot(x,y,z,'o')
    
    # Exibindo o gráfico criado
    plt.show()
```

## Exemplo de pontos de controle ($P_0,  P_1, ..., P_n$)

```python
pontos = [[0, 0, 0], [0, 4, 0], [2, 5, 0], [4, 5, 0], [5, 4, 0], [5, 1, 0], [4, 0, 0], [1, 0, 3], [0, 0, 4], [0, 2, 5], [0, 4, 5], [4, 5, 5], [5, 5, 4], [5, 5, 0]]
plot_bezier_3D(pontos)
```
